```python
def hexToDec(hexNum):
    if len(hexNum) > 3 or bool(search("[g-zG-Z]+",hexNum)):
        return None;
    answ = 0;
    for i, val in enumerate(list(hexNum[::-1])):
        answ += 16**i * hexNumbers[val]
    return answ
```
