```python
class Dog:
    def __init__(self, name="Izabel", legs=4, speak="woof"):
        self.name = name
        self.legs = legs
        self.__speak = speak
        
    def speak(self):
        print(f"{self.name} says: {self.__speak}")
```

```python
a = Dog()
```

```python
a.speak()
```

```python
def factorial(num : int):
    if not isinstance(num, int):
        raise TypeError(f"Provide  int :  num is {num} and {type(num)}")
    elif num != 1:
        return num * factorial(num - 1)
    elif num <= 1:
        return 1
```

```python
factorial(5)
```
